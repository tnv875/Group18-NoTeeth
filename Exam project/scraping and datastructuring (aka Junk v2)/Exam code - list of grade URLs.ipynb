#### Import modules, and create logfile

```python
import pandas as pd
import numpy as np
import requests
import scraping_class
from bs4 import BeautifulSoup

logfile = 'log.csv' ## name of log file.
connector = scraping_class.Connector(logfile)
```

#### Loop over each page, and append to masterlist

```python
list_of_grade_urls = []
url = "http://karakterstatistik.stads.ku.dk/Search/Courses"
for i in range(95):
    data = {
        "searchText": "", 
        "term": "", 
        "block": "", 
        "faculty": "",
        "institute": "null",
        "page": i,
        }
    response = (requests.post(url, data=data))
    html = response.text
    soup = BeautifulSoup(html)
    links = soup.findAll('a')
    list_of_grade_urls.extend([link['href'] for link in links if link.has_attr('href')])
```
